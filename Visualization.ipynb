## 1. Fix Distribution

```python
import os
import csv

from sklearn.utils import shuffle

STEERING_CORRECTION_LEFT  = 0.2
STEERING_CORRECTION_RIGHT = 0.2
USE_SIDE_CAMERAS = True 
FLIP_IMAGES = True

## self defined data structure
class SteeringData:
    def __init__(self, image_path, steer, flipped_flag):
        self.image_path = image_path
        self.steer = steer
        self.flipped_flag = flipped_flag
        self.shadow_flag = 0
        self.bright_flag = 0
        self.blur_flag = 0


## Reading data from csv file 
## saving it as the self defined data structure
def csv_read(csv_path):
    print("Reading data from csv file...")
    data = []
    with open(csv_path) as csv_file:
        reader = csv.reader(csv_file)

        for line in reader:
            center_image_path = line[0]
            if os.path.isfile(center_image_path):
                # The data saved in csv file with the string type
                center_steer = float(line[3])
                data.append(SteeringData(center_image_path, center_steer, 0))
                if FLIP_IMAGES:
                    data.append(SteeringData(center_image_path, -center_steer,1)) 

            if USE_SIDE_CAMERAS:
                left_image_path = line[1]
                if os.path.isfile(left_image_path):
                    # a positive steering value means turn right.
                    left_steer = center_steer + STEERING_CORRECTION_LEFT
                    data.append(SteeringData(left_image_path, left_steer, 0))
                    if FLIP_IMAGES:
                        data.append(SteeringData(left_image_path, -left_steer, 1))
                right_image_path = line[2]
                if os.path.isfile(right_image_path):
                    right_steer = center_steer - STEERING_CORRECTION_RIGHT
                    data.append(SteeringData(right_image_path, right_steer, 0))
                    if FLIP_IMAGES:
                        data.append(SteeringData(right_image_path, -right_steer, 1))
    print("Reading is done.")

    return shuffle(data)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

dataset_path = '/media/ubuntu16/Documents/datasets/CarND/'
csv_path = os.path.join(dataset_path, "driving_log.csv")

# just save the image path to the "training_set" list
# The real image will be feed into the network while training
data = csv_read(csv_path)
training_set, validation_set = train_test_split(data, test_size=0.2)

steers = [item.steer for item in training_set]

# image_number*camera_number*Fliped(1,2)=total_number
print("\nTraining set size: {}".format(str(len(training_set))))
print("\nValidation set size: {}".format(str(len(validation_set))))

print("Max correction: {}".format(np.max(steers)))
print("Min correction: {}".format(np.min(steers)))
```

```python
"""
STEP 1: 画出原始方向角统计直方图
"""

import matplotlib.pyplot as plt
%matplotlib inline

# Split the items into 25 bins

bin_count = 25
max_bin = np.max(steers)
min_bin = np.min(steers)
spread = max_bin - min_bin
bin_size = spread / bin_count

bins = [min_bin + i*bin_size for i in range(bin_count)]
# append last bin
bins.append(max_bin + 0.1)

# plot to visualize the initial distribution
# hist : The array which saves the values of the histogram. 
hist, bin_edges = np.histogram(steers, bins)
# adding value for the last bin 
plt.bar(bin_edges, np.append(hist, 0), width=0.06)
plt.title("Original Distribution")
plt.show
```

```python
'''
STEP 2：将少于均值的个数补齐，高于的不变
方法：计算每个bin内少于均值多少倍，将此bin复制n倍（具体执行时是针对bin内每个数执行的，
     因为每个bin内的数都要复制n倍）
特殊情况：假设此bin内有4个数，当倍数不是整数时，如4.5, 目标是18个数， 每次每个数复制4倍，得到16个数，
     但复制时每个数会结余小数倍，当凑够1时复制一次，因为4个数凑够2次1,某两个数将多复制一次，最后得到
     额外的两个数（0.5倍的bin等价于每隔4*0.5=2个数取1个数）
'''
import copy
import random

desired_per_bin = int(np.mean(hist)*1)
# return an array of copy times for every bin
copy_times = np.float32((desired_per_bin-hist)/hist)

# X is the dataset list
def augment_dataset(x):
    # accumulate the copy times of less one
    copy_times_accum = np.zeros_like(copy_times)
    augmented = []
     
    for i in range(len(x)):
        data = x[i]
        
        index = np.digitize(data.steer, bin_edges) -1
        copy_times_accum[index] += copy_times[index]
        copy_times_integer = np.int32(copy_times_accum[index])
        copy_times_accum[index] -= copy_times_integer
        
        for j in range(copy_times_integer):
            new_data = copy.deepcopy(data)
            new_data.shadow = int(np.random.uniform(0,1) + 0.5)
            new_data.blur   = int(np.random.uniform(0,1) + 0.5)
            new_data.bright = int(np.random.uniform(0,1) + 0.5)
            augmented.append(new_data)
    
    return augmented

augmented = augment_dataset(training_set)
# original data + augmented data
training_set += augmented

print("\nTraining set size: {}".format(str(len(training_set))))
print("\nValidation set size: {}".format(str(len(validation_set))))
```

```python
steers = [item.steer for item in training_set]
bin_count = 25
max_bin = np.max(steers)
min_bin = np.min(steers)
spread = max_bin - min_bin
bin_size = spread / bin_count

bins = [min_bin + i*bin_size for i in range(bin_count)]
bins.append(max_bin + 0.1)

# plot to visualize the initial distribution
# hist : array-The values of the histogram. 
hist, bin_edges = np.histogram(steers, bins)
plt.bar(bin_edges, np.append(hist, 0), width=0.06)
plt.title("Augmented Distribution")
plt.show
```

```python
'''
STEP 3：随机削减掉高出平均值的bin,使数据集大小恢复到初始值
方法：计算每个bin的保留比例（desired/bin_count），然后在(0~1)的均匀分布内随机生成一个值，
     若此值小于此概率，则保留此数据，否则丢弃。（因为是均匀分布，故保留的概率与期望比例大致相等）
结果：最终使得仅有 desired/bin_count 的数据保留，即期望均值个数
'''

keep_percentage = np.float32(desired_per_bin/hist)

# item is a single data in dataset
def should_keep(item):
    prob_to_keep = keep_percentage[np.digitize(item.steer, bin_edges)-1]
    
    random_prob = np.random.uniform(0,1)
    
    return (random_prob <= prob_to_keep)

trimmed_training_set = [item for item in training_set if should_keep(item)]

print("Training set size: {}".format(str(len(trimmed_training_set))))

steers = [item.steer for item in trimmed_training_set]
hist, bin_edges = np.histogram(steers, bins)
plt.bar(bin_edges, np.append(hist, 0), width=0.06)
plt.title("Distribution Corrected")
plt.show
```

```python
from os import listdir
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import copy

dataset_path = '/media/ubuntu16/Documents/datasets/CarND/'
images_path = dataset_path + 'IMG/'
image_path = images_path + listdir(images_path)[0]

# default colorspace loaded in opencv is BGR
# but plt will index them by the RGB order
image = cv2.imread(image_path) # BGR
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

fig1, ax1 = plt.subplots(1, 3, figsize = (12, 4))
ax1[0].imshow(image)
ax1[0].set_title("BGR")
ax1[1].imshow(image_rgb)
ax1[1].set_title("RGB")
ax1[2].imshow(image_hsv)
ax1[2].set_title("HSV")

# grab the brightness channel
# note that image's dtype is np.uint8, forcing type transform will change the number value!
v_channel = image_hsv[:,:,2]

# get a random number to represent the change in brightness
# note that all the v channel values should add a same number
'''
int8: -128~127, int16: -32768~32767, uint8: 0~255
differnt types of data can not operate with each other 
'''
brightness_change = np.random.randint(0,100,dtype=np.uint8)


# apple brightness change
# be sure that v_channel value will not less 0 or great 255 while adding or substracting !
if (np.random.uniform(0,1) > 0.5):
    v_channel[v_channel>(255-brightness_change)] = 255
    v_channel[v_channel<=(255-brightness_change)] += brightness_change
    
else:
    v_channel[v_channel<(brightness_change)] = 0
    v_channel[v_channel>=(brightness_change)] -= brightness_change
# using v[v >= make_positive] + brightness_change can avoid this problem!

# put the changed v channel back to hsv and then covert to rgb
# this line can be deleted becase that v_channel is the reference of image_hsv[]
image_hsv[:,:,2] = v_channel

bgr = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2BGR)
rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)

fig = plt.figure(figsize = (12, 4))
#fig.suptitle("Randomly changing the brightness of an image", y = 0) #, fontsize=12
#fig2, ax2 = plt.subplots(1, 3, figsize = (12, 4))
ax = plt.subplot(131)
ax.imshow(bgr)
ax.set_title("BGR")

ax = plt.subplot(132)
ax.set_title("RGB")
ax.imshow(rgb)

ax = plt.subplot(133)
ax.set_title("HSV")
ax.imshow(image_hsv)

plt.show()

```

## 3. Random Shadow

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset_path = '/media/ubuntu16/Documents/datasets/CarND/'
images_path = dataset_path + 'IMG/'

image_path = images_path + os.listdir(images_path)[1]
image = cv2.imread(image_path)

rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.title("original")
plt.show()
```

```python
# !!
height, width = image.shape[:2]

number_of_shadows = np.random.randint(1, 6)
list_of_shadows = []

# define each shadow by randomly determine 
# the number and positions of a polygon's vertices
for i in range(number_of_shadows):
    shadow_vertices = []
    
    number_of_vertices = np.random.randint(3,26)
    
    for j in range(number_of_vertices):
        position_x = width * np.random.uniform()
        position_y = height * np.random.uniform()
        
        shadow_vertices.append((position_x,position_y))
    
    list_of_shadows.append(np.array([shadow_vertices], dtype=np.int32))
        
# create a mask with the same dimensions as the original image
mask = np.zeros((height,width))

# fill all the shadow polygon
for shadow in list_of_shadows:
    cv2.fillPoly(mask, shadow, 255)
    
plt.imshow(mask, cmap='gray')
plt.title("shadow")
plt.show()
    
```

```python
print(len(list_of_shadows))
print(list_of_shadows[0].shape)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
v_channel = hsv[:,:,2]

# randomly choose a drakness of the shadows
# lower numbers result in darker shadows

random_darkness = np.random.randint(45, 75) /100.0

# indexed by array
v_channel[mask==255] = v_channel[mask==255]*random_darkness

rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

plt.imshow(rgb)
plt.title("shadow")
plt.show()
```

## 4. Visualize Cropping2D

```python
from keras import backend
from keras.models import Sequential
from keras.layers import Cropping2D

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset_path = '/media/ubuntu16/Documents/datasets/CarND/'
filename = os.listdir(dataset_path + 'IMG/')[2]

image = cv2.imread(dataset_path+'IMG/'+filename)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

model = Sequential()

model.add(Cropping2D(cropping=((50, 20),(0,0)), input_shape=(160,320,3)))

# output the intermediate layer! 
## can be used to show learned features for any layer
layer_output = backend.function([model.layers[0].input], [model.layers[0].output])

# note that the shape of the image suited for the network!
cropped_image = layer_output([image[None,...]])[0][0]

# convert to uint8 for visualization
cropped_image = np.uint8(cropped_image)


f1, (ax1,ax2) = plt.subplots(1, 2, figsize=(24, 9))
f1.tight_layout()
ax1.imshow(image)
ax1.set_title('Orig: '+ str(image.shape), fontsize=25)
ax2.imshow(cropped_image)
ax2.set_title('cropped: '+ str(cropped_image.shape), fontsize=25)
plt.show()

'''
this function runs the computation graph that we have created in the code, 
taking input from the first parameter and extracting the number of outputs 
as per the layers mentioned in the second parameter. 
'''
```
